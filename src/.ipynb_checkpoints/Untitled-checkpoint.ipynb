```python
import random
import math
import torch




class Neuron:
        
    def __init__(self, nin):
        self.w = [torch.tensor(random.uniform(-1,1), requires_grad=True) for _ in range(nin)]
        self.b = torch.tensor(random.uniform(-1.0,1.0), requires_grad=True)

    def parameters(self):
        return self.w

    def __call__(self, x):
        act = sum((wi*xi for wi,xi in zip(self.w, x)), self.b)
        return torch.tanh(act)

class Layer:

    def __init__(self, nin, nout):
        self.neurons = [Neuron(nin) for _ in range(nout)]

    def __call__(self, x):
        out = [n(x) for n in self.neurons]
        return out[0] if len(out) == 1 else out

    def parameters(self):
        return [p for n in self.neurons for p in n.parameters()]

    def __repr__(self):
        return f"Layer of [{', '.join(str(n) for n in self.neurons)}]"

class MLP:

    def __init__(self, nin, nouts):
        sz = [nin] + nouts
        self.layers = [Layer(sz[i], sz[i+1]) for i in range(len(nouts))]

    def __call__(self, x):
        for layer in self.layers:
            x = layer(x)
        return x

    def parameters(self):
        return [p for layer in self.layers for p in layer.parameters()]

    def __repr__(self):
        return f"MLP of [{', '.join(str(layer) for layer in self.layers)}]"
```

```python
# [weight(g), sweetness(1-10)] → fruit
x = [
    [150, 8],   # apple
    [120, 9],   # apple  
    [180, 7],   # apple
    [300, 3],   # orange
    [280, 4],   # orange
    [320, 2],   # orange
    [500, 6],   # watermelon
    [600, 7],   # watermelon
    [550, 5],   # watermelon
]
y = [0, 0, 0, 1, 1, 1, 2, 2, 2] 

xs = torch.tensor(x).float()
ys = torch.tensor(y)

n = MLP(2,[4,3])
```

```python
# Loss function
def softmax(outputs):
    ex = [val.exp() for val in outputs]
    s = sum(ex)
    return [val/s for val in ex]

for i in range(10):
    learning_rate = 1
    ypred = [torch.stack(softmax(n(j))) for j in xs]
    loss = 0
    for j in range(len(xs)):
        probs = softmax(n(xs[j]))
        correct_class = ys[j].item()
        loss += -probs[correct_class].log()

    loss = loss / len(xs) 

    for w in n.parameters():
        w.grad =None
    loss.backward()
    for w in n.parameters():
        w.data -= learning_rate * w.grad    
    print(loss)
    
```

```python
print(softmax(n([160, 8]))) # should be apple (0)
print(softmax(n([290, 3])))   # should be orange (1)
print(softmax(n([520, 6])))  # should be watermelon (2)
```
